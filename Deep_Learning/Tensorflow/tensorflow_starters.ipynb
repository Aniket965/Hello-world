### Variables

````python
'''Ordinary Variable initialization'''
a = tf.Variable(2,name= 'scalar' )
b = tf.Variable([2,3],name= 'matrix' )
c = tf.Variable(tf.zeros([800,10]))
sess.run(a.initializer)
sess.run(b.initializer)
sess.run(c.initializer)

print(sess.run(a),'\n')
print(sess.run(b),'\n')
print(sess.run(c),'\n')
'''The tf.Variables contains many ops'''
````

````python
'''Variable initializer with get variables'''
s = tf.get_variable("scalar", initializer=tf.constant(2)) 
m = tf.get_variable("matrix", initializer=tf.constant([[0, 1], [2, 3]]))
W = tf.get_variable("big_matrix", shape=(784, 10), initializer=tf.zeros_initializer())
sess.run(tf.global_variables_initializer())
#sess.run(tf.variables_initializer([s, m])) #->needed variable initialization
# sess.run(W.initializer) #-> single variable initializer then to get data do W.eval()
print(sess.run(s))
print(sess.run(m))
print(sess.run(W))
'''This variables cannot be reinitialized'''
````

````python
'''tf.assign'''
W = tf.Variable(10)
assign_op = W.assign(100)
with tf.Session() as sess:
    sess.run(W.initializer)
    sess.run(assign_op)
    print(W.eval())
````

````python
'''tf.assign'''
W = tf.Variable(10)
assign_op = W.assign(100 * W)
with tf.Session() as sess:
    sess.run(W.initializer)
    print(sess.run(assign_op))
    print(sess.run(assign_op))
    print(sess.run(assign_op))
    print(sess.run(assign_op))
    print(sess.run(assign_op))
#     print(W.eval())
````

````python
'''assign_add and assign_sub'''
my_var = tf.Variable(10)
with tf.Session() as sess:
    sess.run(my_var.initializer)
    print(sess.run(my_var.assign_add(10))) # >> 20
    print(sess.run(my_var.assign_add(10))) # >> 30
    print(sess.run(my_var.assign_sub(2))) # >> 18
````

````python
'''Control dependencies'''
a = tf.Variable(2,name= 'scalar' )
b = tf.Variable(2,name= 'scalar' )
sess =  tf.Session() 
sess.run(a.initializer)
sess.run(b.initializer)
g = tf.get_default_graph()
with g.control_dependencies([a,b]):
    d = tf.add(a,b)
    print(sess.run(d))
    '''This will be executed after the a and b is executed'''
````

### Placeholders

````python
'''Placeholder are used so that we can assign unknown values'''
a = tf.placeholder(tf.float32,name='a')
b = tf.Variable([5,5,5],name = 'b',dtype = tf.float32)
c = tf.add(a,b)
sess.run(a.initializer)
with tf.Session() as sess:
    sess.run(b.initializer)
    print(sess.run(c,feed_dict={a: [1,1,1]})) # we provide values for palceholder using the feed dict
    
````

````python
'''check if the tensor is feedable'''
g = tf.get_default_graph()
g.is_feedable(a)
````

````python
'''we can override the operation by giving the feed dict value'''
a = tf.add(2,3)
b = tf.multiply(a,5)
with tf.Session() as sess:
    print(sess.run(b))
    print(sess.run(b,feed_dict={a:2}))
````

### Lazy loading  
`Defer creating/initializing an object 
until it is needed
`

##### `Normal loading`

````python
x = tf.Variable(10, name='x')
y = tf.Variable(20, name='y')
z = tf.add(x, y)
writer = tf.summary.FileWriter('tmp/graph/normal_loading', tf.get_default_graph())
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    for _ in range(10):
        sess.run(z)
writer.close()
````

#### `Lazy loading`

````python
x = tf.Variable(10, name='x')
y = tf.Variable(20, name='y')

writer = tf.summary.FileWriter('tmp/graph/lazy_loading', tf.get_default_graph())
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    for _ in range(10):
        sess.run(tf.add(x, y)) # someone decides to be clever to save one line of code
writer.close()
````
