```python
#Sampling from a gaussian distribution
r=np.random.randn(10000,2)
plt.scatter(r[:,0], r[:,1])
plt.axis('equal')
plt.show()
```

```python
r[:,1]= 5*r[:,1]+2
plt.scatter(r[:,0], r[:,1])
plt.axis('equal')
plt.show()
```

```python
#Sampling from a general multivariate normal
#variance of 1 on first dimension, 3 on second, covariance b/w them is 0.8 
cov=np.array([[1,0.8],[0.8,3]])
```

```python
from scipy.stats import multivariate_normal as mvn
```

```python
mu =np.array([0,2])
```

```python
r=mvn.rvs(mean=mu, cov=cov, size=1000)
```

```python
plt.scatter(r[:,0], r[:,1])
plt.axis('equal')
plt.show()
```

```python
r = np.random.multivariate_normal(mean=mu, cov=cov, size=1000)
```

```python
plt.scatter(r[:,0], r[:,1])
plt.axis('equal')
plt.show()
```
